## while loop

```python
## cat sounds "meow" three times . write a code
```

```python
i = 3
while i != 0 :
    print("meow")
    i -= 1
```

```python
i = 0 
while i < 3:
    print("meow")
    i += 1
```

## dictionaries

```python
students = {"pallavi":"machiriyal","sai":"kandukur","yaswanth":"ongole"}# little clumsy to read
print(students)
```

```python
students = {"pallavi":"machiriyal",
            "sai":"kandukur",
            "yaswanth":"ongole"}#we can write like this also
print(students["yaswanth"])
```

```python
students = {"pallavi":"machiriyal",
            "sai":"kandukur",
            "yaswanth":"ongole"}#we can write like this also
for student in students:
    print(student,students[student],sep=",")
```

```python
students = {"pallavi":"machiriyal",
            "sai":"kandukur",
            "yaswanth":"ongole"}#we can write like this also
for student in students:
    print(student)
```

```python
## now i have students name,house,patronus.
students = [
    {"name":"yaswanth","house":"plp","patronus":"venkateswarlu"},
    {"name":"pallavi","house":"mcrl","patronus":None},
    {"name":"sai","house":"ayprp","patronus":"mk"}   
]
for student in students:
    print(student["name"],student["house"],student["patronus"],sep = ",")
```
