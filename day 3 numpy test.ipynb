## Section 6 — Aggregations & the `axis` Argument

**Q22.** Given
```python
M = np.array([[3, 7, 2, 8],
              [5, 1, 9, 4],
              [6, 0, 3, 2]])
```
Compute:
- (a) total sum of all elements
- (b) sum of each column
- (c) sum of each row
- (d) maximum of each row
- (e) minimum of each column

````python
import numpy as np

M = np.array([[3, 7, 2, 8],
              [5, 1, 9, 4],
              [6, 0, 3, 2]])

print(np.sum(M))
print(np.sum(M, axis=0))
print(np.sum(M, axis=1))
print(np.max(M, axis=1))
print(np.min(M, axis=0))
````

**Q27.** Generate a 1-D array of 20 random numbers drawn from a standard normal distribution (seed = 0). Then compute its mean, std, min, and max.

**Q28.** Normalize the columns of a 4×3 random matrix so that each column has mean 0 and standard deviation 1. (Use broadcasting — no loops.)

````python
import numpy as np

rng = np.random.default_rng(0)
M = rng.random((4, 3))
normalized = (M - M.mean(axis=0)) / M.std(axis=0)
print(normalized)
print(normalized.mean(axis=0))
print(normalized.std(axis=0))
````

**Q29.** Given marks of 5 students in 3 subjects:
```python
marks = np.array([[78, 82, 90],
                  [65, 70, 72],
                  [88, 95, 92],
                  [54, 60, 58],
                  [80, 85, 87]])
```
Find: (a) each student's average, (b) each subject's average, (c) the index of the top-scoring student overall (highest total), (d) the subject in which the class performed best on average.

````python
import numpy as np

marks = np.array([[78, 82, 90],
                  [65, 70, 72],
                  [88, 95, 92],
                  [54, 60, 58],
                  [80, 85, 87]])

student_avg = marks.mean(axis=1)
subject_avg = marks.mean(axis=0)
top_student = np.argmax(marks.sum(axis=1))
best_subject = np.argmax(subject_avg)

print(student_avg)
print(subject_avg)
print(top_student)
print(best_subject)
````
